```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 3.1 Выбор источника и инструментальных средств для получения данных
# Обсуждение источников данных и инструментов, используемых для сбора и обработки данных.

# 3.2 Подготовка данных
data = pd.read_csv('crime.csv')
data.dropna(inplace=True)  # Очистка данных от пропусков

# 3.3 Анализ и визуализация полученных данных
# Загрузка данных о преступлениях
data = pd.read_csv('crime2.csv', encoding='utf-8')

# Преобразование столбца 'month' в формат даты
data['month'] = pd.to_datetime(data['month'], format='%d.%m.%Y')

# Сгруппировать данные по годам и вычислить суммарное количество преступлений за каждый год
data['Year'] = data['month'].dt.year
yearly_data = data.groupby('Year')['Total_crimes'].sum()

# Построение графика общего количества преступлений по годам
plt.figure(figsize=(10, 6))
plt.plot(yearly_data.index, yearly_data.values, marker='o', color='b', linestyle='-')
plt.title('Общее количество преступлений по годам')
plt.xlabel('Год')
plt.ylabel('Общее количество преступлений')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Построение графиков для различных типов преступлений
plt.figure(figsize=(12, 6))
plt.plot(data['month'], data['Serious'], label='Серьезные преступления', marker='o', linestyle='-')
plt.plot(data['month'], data['Huge_damage'], label='Большой ущерб', marker='o', linestyle='-')
plt.plot(data['month'], data['Ecological'], label='Экологические преступления', marker='o', linestyle='-')
plt.plot(data['month'], data['Terrorism'], label='Терроризм', marker='o', linestyle='-')
plt.plot(data['month'], data['Extremism'], label='Экстремизм', marker='o', linestyle='-')
plt.plot(data['month'], data['Murder'], label='Убийства', marker='o', linestyle='-')
plt.plot(data['month'], data['Harm_to_health'], label='Причинение вреда здоровью', marker='o', linestyle='-')
plt.plot(data['month'], data['Rape'], label='Изнасилования', marker='o', linestyle='-')
plt.plot(data['month'], data['Theft'], label='Кражи', marker='o', linestyle='-')
plt.plot(data['month'], data['Vehicle_theft'], label='Кражи транспортных средств', marker='o', linestyle='-')
plt.plot(data['month'], data['Fraud_scam'], label='Мошенничество', marker='o', linestyle='-')
plt.plot(data['month'], data['Hooligan'], label='Хулиганство', marker='o', linestyle='-')
plt.plot(data['month'], data['Drugs'], label='Наркотики', marker='o', linestyle='-')
plt.plot(data['month'], data['Weapons'], label='Оружие', marker='o', linestyle='-')
plt.legend(loc='upper right')
plt.title('Различные типы преступлений по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество преступлений')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Построение матрицы корреляции между различными типами преступлений с русскими подписями
crime_types = ['Serious', 'Huge_damage', 'Ecological', 'Terrorism', 'Extremism', 'Murder', 'Harm_to_health', 'Rape', 'Theft', 'Vehicle_theft', 'Fraud_scam', 'Hooligan', 'Drugs', 'Weapons']
crime_types_ru = ['Серьезные', 'Большой ущерб', 'Экологические', 'Терроризм', 'Экстремизм', 'Убийства', 'Вред здоровью', 'Изнасилования', 'Кражи', 'Кражи ТС', 'Мошенничество', 'Хулиганство', 'Наркотики', 'Оружие']

correlation_matrix = data[crime_types].corr()
correlation_matrix.columns = crime_types_ru
correlation_matrix.index = crime_types_ru

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Матрица корреляции между различными типами преступлений')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных о преступлениях
data = pd.read_csv('crime2.csv', encoding='utf-8')

# Преобразование столбца 'month' в формат даты и выделение месяца и года
data['month'] = pd.to_datetime(data['month'], format='%d.%m.%Y')
data['Month'] = data['month'].dt.month
data['Year'] = data['month'].dt.year

# Разделение данных на признаки и целевую переменную
X = data.drop(['Serious', 'month'], axis=1)  # Используем все столбцы кроме 'Serious' и 'month' как признаки
y = data['Serious']  # 'Serious' будет целевой переменной

# Стандартизация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Разделение данных на обучающую и тестовую выборку
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Выбор модели и метрик оценки
# Использование логистической регрессии для классификации с увеличением количества итераций
model = LogisticRegression(max_iter=2000)
model.fit(X_train, y_train)  # Обучение модели

# Предсказание на тестовой выборке
predictions = model.predict(X_test)

# Оценка модели
print(classification_report(y_test, predictions, zero_division=0))

# Визуализация важности признаков
feature_importance = pd.DataFrame(model.coef_[0], index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

# Создание словаря для перевода названий признаков
feature_translation = {
    'Total_crimes': 'Всего преступлений',
    'Huge_damage': 'Огромный ущерб',
    'Ecological': 'Экологические',
    'Terrorism': 'Терроризм',
    'Extremism': 'Экстремизм',
    'Murder': 'Убийства',
    'Harm_to_health': 'Вред здоровью',
    'Rape': 'Изнасилование',
    'Theft': 'Кража',
    'Vehicle_theft': 'Кража транспорта',
    'Fraud_scam': 'Мошенничество',
    'Hooligan': 'Хулиганство',
    'Drugs': 'Наркотики',
    'Weapons': 'Оружие',
    'Month': 'Месяц',
    'Year': 'Год'
}

# Перевод индексов признаков
feature_importance.index = feature_importance.index.map(feature_translation)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importance.importance, y=feature_importance.index)
plt.title('Важность признаков')
plt.xlabel('Важность')
plt.ylabel('Признаки')
plt.show()
```
